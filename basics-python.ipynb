```python
def phan_tich_bong_da(skill1, salary1, skill2, salary2):
    """
    Print job titles with their corresponding average salaries using explicitly named parameters.

    :param skill1: The first job title or skill.
    :param salary1: The average salary for the first job title or skill.
    :param skill2: The second job title or skill.
    :param salary2: The average salary for the second job title or skill.
    """
    print('Cầu thủ bóng đá và mức lương trung bình:')
    print(f' - {skill1}: ${salary1}')
    print(f' - {skill2}: ${salary2}')

phan_tich_bong_da(
    skill1='Cầu thủ chuyên nghiệp', salary1=900000,skill2='Cầu thủ nghiệp dư', salary2=1000)
```

```python
def phan_tich_bong_da(*đối_số):
    """
    Print skills with their corresponding average salaries using positional arguments.

    Each argument is expected to be a tuple containing a job title or skill and its average salary.

    :param args: A sequence of tuples, where each tuple contains a job title or skill and its average salary.
    """
    print('Cầu thủ bóng đá và mức lương trung bình:')
    for tien_luong in đối_số:
        cau_thu, muc_luong = tien_luong
        print(f'- {cau_thu}: ${muc_luong:,}')

phan_tich_bong_da(('Cầu thủ chuyên nghiệp', 900000), ('Cầu thủ bán chuyên', 30000), ('Cầu thủ nghiệp dư', 7000))
     
```

```python
%%writefile phan_tich.py

def tinh_muc_luong(muc_luong_co_ban, bonus=.1):
    """
  Tính tổng mức lương dựa vào mức lương cơ bản và mức thưởng

  Args:
    base_salary (float): The base salary.
    bonus_rate (float): The bonus rate. Default is .1.
  
  Returns:
    float: The total salary.
    """
    return muc_luong_co_ban * (1 + bonus)

def tinh_bonus(tong_muc_luong, muc_luong_co_ban):
    """
  Tính mức thưởng dựa vào tổng mức lương và mức lương cơ bản.

  Args:
    total_salary (float): The total salary.
    base_salary (float): The base salary.

  Returns:
    float: The bonus rate.
    """
    return (tong_muc_luong - muc_luong_co_ban) / muc_luong_co_ban
```

```python
class DanhSachCuaHuy:
    def __init__(self):
        """Khởi tạo một danh sách trống."""
        self._things = []

    def add(self, thing):
        """Thêm một mục vào cuối danh sách."""
        self._things.append(thing)

    def __getitem__(self, index): 
        """Truy xuất mục theo chỉ mục."""
        return self._things[index]
    
    def __setiem__(self, index, value):
        """Đặt một mục tại một chỉ mục cụ thể."""
        self.things[index] = value

    def __repr__(self):
        """Trả về một chuỗi biểu diễn danh sách."""
        return str(self._things)
   
    def __len__(self):
        """Trả về độ dài của danh sách."""
        return len(self._things)
```

```python
def calculate_salary(base_salary, bonus_rate=.1):
    return base_salary * (1 + bonus_rate)

def calculate_bonus(total_salary, base_salary):
    return (total_salary - base_salary) / base_salary
```

```python
class BaseSalary:
    def __init__(self, base_salary, bonus_rate=0.1, symbol="$"):
        self.base_salary = base_salary
        self.bonus_rate = bonus_rate
        self.symbol = symbol
        self.total_salary = base_salary * (1 + bonus_rate)
        self.bonus = self.total_salary - self.base_salary

    def __repr__(self):
        return f'{self.symbol}{self.base_salary:,.0f}'
    
    def show_salary(self):
        return f'{self.symbol}{self.total_salary:,.0f}'

    def show_bonus(self):
        return f'{self.symbol}{self.bonus:,.0f}'
                
                 
```
